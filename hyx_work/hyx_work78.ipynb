## 题7-2

```python
import pandas as pd
from pyecharts import options as opts
import random

df = pd.read_csv(r'C:\Users\admin\Desktop\hyx数据可视化作业\sales.csv',header=None)

df_1 = df.groupby(df[11]).sum()

labels = df_1._stat_axis.values.tolist()

x_data = labels
y_data = df_1[16]
# 将数据转换为列表加元组的格式（[(key1, value1), (key2, value2)]）
data=[list(z) for z in zip(x_data, y_data)]
# 数据排序
data.sort(key=lambda x: x[1])

# 创建饼形图并设置画布大小
pie=Pie(init_opts=opts.InitOpts(width='1000px',height='724px'))
# 为饼形图添加数据
pie.add(
 series_name="地区", # 序列名称
 data_pair=data, # 数据
 radius=["8%","90%"], #内外半径
 center=["60%","60%"], # 位置
 rosetype='area', # 玫瑰图
 color='auto' # 颜色自动渐变
 )
pie.set_global_opts(
 # 不显示图例
 legend_opts=opts.LegendOpts(is_show=False),
 # 视觉映射
 visualmap_opts=opts.VisualMapOpts(is_show=False,
 min_=100, # 颜色条最小值
 max_=4500000, # 颜色条最大值
 )
)
pie.set_series_opts(
 # 序列标签
 label_opts=opts.LabelOpts(position='inside', # 标签位置
 rotate=45,
  font_size=11) # 字体大小
 #formatter="{b}: {c}") # 标签格式
 )

pie.render_notebook()
```

## 8-2 GDP

```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Map

# 获得数据源之后直接在这个地方进行修改即可
provinces = ["广东", "北京", "上海", "辽宁", "湖南", "四川", "西藏"]
guangdong_city = ["汕头市", "汕尾市", "揭阳市", "阳江市", "肇庆市", "广州市", "惠州市"]
country = ["China", "Canada", "Brazil", "Russia", "United States", "Africa", "Germany"]
value = [300, 100, 2000, 800, 10000, 400, 5000]
```

```python
# 显示其中的某些省市和数据
def map_base() -> Map:
    c = (
        Map()
        .add("", [list(z) for z in zip(provinces, value)], "china")
        .set_global_opts(title_opts=opts.TitleOpts(title="map-基本图形"))
    )
    return c
if __name__ == '__main__':
    city_map = map_base()
    city_map.render(path="test_map_1.html")
```

```python
# 连续性数据显示，不同颜色不同省份
def map_visualmap() -> Map:
    c = (
        Map()
        .add("", [list(z) for z in zip(provinces, value)], "china")
        .set_global_opts(
            title_opts=opts.TitleOpts(title="China_data"),
            visualmap_opts=opts.VisualMapOpts(max_= 2000),
        )
    )
    return c

city_ = map_visualmap()
city_.render_notebook()
```

## 8-3 世界新冠数据

```python
# 显示世界地图
def map_world() -> Map:
    c = (
        Map()
        .add("", [list(z) for z in zip(country, value)], "world")
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="世界新冠数据",subtitle="截止2021-06-26"),
            visualmap_opts=opts.VisualMapOpts(max_=2000),
        )
    )
    return c

country_ = map_world()
country_.render_notebook()
```
