```python
import pandas as pd
from pyecharts import options as opts
import random

df = pd.read_csv(r'C:\Users\admin\Desktop\hyx数据可视化作业\sales.csv',header=None)

df_1 = df.groupby(df[2]).sum()

# 转化之后分别对应16 19

x_labels = df_1._stat_axis.values.tolist()

from pyecharts.charts import Bar
from pyecharts.charts import Line

y_17 = list(df_1[16])
y_17 = [ round(i,1) for i in y_17 ]
y_20 = list(df_1[19])
y_20 = [ round(i,1) for i in y_20 ]



def bar_line_combine_with_two_axis():
    bar = Bar(init_opts=opts.InitOpts(theme='light',
                                      width='1000px',
                                      height='600px'))
    bar.add_xaxis(x_labels)
    # 添加一个Y轴
    bar.extend_axis(yaxis=opts.AxisOpts())
    bar.add_yaxis('销售额',y_17, yaxis_index=0)

    line = Line(init_opts=opts.InitOpts(theme='light',
                                        width='1000px',
                                        height='600px'))
    line.add_xaxis(x_labels)
    # 将line数据通过yaxis_index指向后添加的Y轴
    line.add_yaxis('利润额',y_20, yaxis_index=1)

    bar.overlap(line)
    return bar


chart = bar_line_combine_with_two_axis()
chart.render_notebook()
```
